```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_figure(images, label):
    # dimension: 2, 108, 108
    images = np.hstack((images[0], images[1]))
    plt.figure()
    plt.imshow(images, cmap="gray")
    if type(label) == str:
        title = label
    else:
        title = str(label) + ": {}".format(CATEGORY_MAPPING[label])
    plt.title(title)
```

## Plot Images from Directory

```python
cat = "plane"
image_fname = "image_test_01_plane_29133.npy"
fname = "/mnt/fs5/nclkong/datasets/norb/test_npy/{}/{}".format(cat, image_fname)
v = np.load(fname)
plot_figure(v, cat) # TODO: Make the label making more flexible
```

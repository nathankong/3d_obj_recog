```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import pickle
```

```python
def load_results_pkl(fname):
    res = pickle.load(open(fname, 'rb'), encoding="latin1")
    return res

def plot_loss_iter(res, train=True):
    assert train
    plt.figure()
    
    if train:
        plt.plot(res["train"]["iter_losses"])
        plt.title("Train Losses")

    plt.title("Loss")
    plt.xlabel("Iteration")

def plot_acc_iter(res, train=True):
    assert train
    plt.figure()
    
    if train:
        plt.plot(res["train"]["iter_top1_accs"], label="Top 1", color="blue")
        plt.title("Train Accuracy")
        
    plt.xlabel("Iteration")
    plt.legend()

def plot_loss_epoch(res, train=True):
    plt.figure()
    
    if train:
        plt.plot(res["train"]["losses"])
        plt.title("Train Losses")

    plt.title("Loss")
    plt.xlabel("Epoch")

def plot_acc_epoch(res, train=True):
    plt.figure()
    
    if train:
        plt.plot(res["train"]["top1_accs"], label="Top 1", color="blue")
        plt.title("Train Accuracy")
    else:
        plt.plot(res["val"]["top1_accs"], label="Top 1", color="blue")
        plt.title("Validation Accuracy")
        print("Max Top-1 Validation Accuracy {}".format(np.max(res["val"]["top1_accs"])))
        
        print(res["val"]["top1_accs"])

    plt.xlabel("Epoch")
    plt.legend()
```

```python
fname = "/mnt/fs5/nclkong/trained_models/norb/test/results.pkl"
res = load_results_pkl(fname)
plot_acc_iter(res)
plot_acc_epoch(res)
plot_acc_epoch(res, train=False)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import pickle
```

```python
def load_results_pkl(fname):
    res = pickle.load(open(fname, 'rb'), encoding="latin1")
    return res

def plot_losses(losses):
    plt.figure()
    plt.plot(losses["train"], label="train")
    plt.plot(losses["val"], label="val")
    plt.title("Loss")
    plt.xlabel("Epoch")
    plt.legend()
    print("Best epoch val loss:", np.min(losses["val"]))

def plot_accs(accs):
    plt.figure()
    plt.plot(accs["train"], label="train")
    plt.plot(accs["val"], label="val")
    plt.title("Accuracy")
    plt.xlabel("Epoch")
    plt.legend()
    print("Best epoch val accuracy:", np.max(accs["val"]))
```

## EXP 01: 40 filters

```python
exp_id = "exp01"
dataset = "norb"
results_fname = "results.pkl"
results_path = "/mnt/fs5/nclkong/trained_models/{}/{}/".format(dataset, exp_id)

res = load_results_pkl(results_path + results_fname)
losses = res["losses"]
accs = res["accs"]

plot_losses(losses)
plot_accs(accs)
```
